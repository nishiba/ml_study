## 列を追加・削除する。

```python
data['new'] = np.random.normal(0, 1, data.shape[0])
data.head()
```

```python
d0 = data.drop(['black', 'hisp', 'married', 'nodegr', 're74', 're75', 're78', 'u74', 'u75', 'treat', 'new'], axis=1)
d0.head()
```

## 行を追加・削除する

```python
d1 = d0.append(pd.DataFrame({'age': [1, 2], 'educ': [2, 3]}))
d1.tail()
```

```python
d1.drop(0).tail()
```

## indexを振り直す。

```python
d2 = d1.reset_index(drop=True)
d2.head()
```

## sortする
