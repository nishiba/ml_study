```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import nsb
```

# 3.3 外れ値の検出

```python
nhanes = pd.read_csv('./pandas_usage/nhanes.csv', sep=',', index_col=0)
```

```python
iris_data = datasets.load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = pd.Series(iris_data.target, dtype='category')
iris.head(2)
```

## 3.3.1 外れ値とは

```python
iris.plot(kind='hist', bins=50, subplots=True);
plt.show()
```

```python
fig, axes = plt.subplots(3, 1)
sns.boxplot(data=iris[iris['species'] == 0], ax=axes[0])
sns.boxplot(data=iris[iris['species'] == 1], ax=axes[1])
sns.boxplot(data=iris[iris['species'] == 2], ax=axes[2])
axes[0].set_ylabel(0)
axes[1].set_ylabel(1)
axes[2].set_ylabel(2)
plt.show()
```
